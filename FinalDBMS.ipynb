```python
PK = {}
FD = []

def check(list1,list2):
    num = 0
    dic= {}
    for i in list1:
        if i in dic:
            if dic[i] == list2[num]:
                num += 1
                continue
            return 1
        dic[i] = list2[num]
        num +=1

for rowName,columnVals in Tabledict.items():
    for rowName2,columnVals2 in Tabledict.items():
        if rowName == rowName2:
            continue
        if (not check(columnVals,columnVals2)):
            FD.append([rowName,rowName2])
FD

```
